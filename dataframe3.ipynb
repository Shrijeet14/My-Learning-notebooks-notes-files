### Groupby Attributes and Methods

```python
# len() --> To find the total number of groups.
# size() --> To find no. of items in each item in a column.
# first() / last() --> To find the first/last item of each group formed by groupby().
# nth() --> To find the nth item of each group formed by groupby().
# get_group() --> After applying the groupby() , it forms a group for all the items diff present , if u want to use a particularly group formed by groupby , we use this.
# groups() --> Gives us a dictionary having all the unique elements as key and their index number as value in the form of list.
# describe() --> It give mathematical summarization of each item of the groupby for every numerical coulmn .
# agg method() --> We can apply multiple aggregate functions on multiple columns . It gives the way to apply different aggregate method on different columns.
# LOOPING ON A GROUP 
#SPLIT - APPLY - COMBINE
```

### agg method()

```python
# agg method
# passing dict
movies.groupby('Genre').agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# Adding both the syntax
movies.groupby('Genre').agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

### Split , apply , combine strategy 

```python
# We use groupby to split .
# Then we use apply to make a custom function and apply it in the groups formed.
# Then we combine them again back .
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
  
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie of each category according to IMDB score
movies.head(2)
```

```python
genres = movies.groupby('Genre')

def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
    
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
# Q5 => Create a function that can return the highest score of any batsman.

temp_df = ipl[ipl['batsman']=='V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).iloc[0]


def highestScore(name): 
    temp_df = ipl[ipl['batsman']==name]
    return (temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).iloc[0])

```

```python
highestScore('RG Sharma')
```

```python
highestScore('MS Dhoni')
```
