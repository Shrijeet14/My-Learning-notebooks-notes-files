## Merge

```python
# there are 4 types :-
# 1) Inner joining .
# 2) outter joining .
# 3) left joining .
# 4) Right joining .
```

### inner joining 

```python
# here we join on the basis of some common column.
# When we will join then only the student ids present in both the table will be selected.
```

```python
students
```

```python
reg
```

```python
students.merge(reg , how='inner' , on="student_id")
```

### Left join

```python
# here also we join on the basis of some common column.
# it joins common items but also all the items present in left table. Left table will be the table on which u will run merge() command and right one will be the one that will come as argument.
```

```python
reg
```

```python
courses
```

```python
courses.merge(reg , how='left' , on = 'course_id') 
# courses --> left table  
# reg --> right table 
```

```python
courses.merge(reg , how='left' , on = 'course_id').sort_values('student_id')
```

### right join

```python
# here also we join on the basis of some common column.
# it joins common items but also all the items present in right table. right table will be the one that will come as argument.
```

```python
# right join
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.tail()
```

```python
students.merge(reg,how='right',on='student_id')
```

### outter join

```python
# here we join on the basis of some common column.
# When we will join then all the columns present in both or in one will be considered .
```

```python
students.merge(reg , how='outer' , on='student_id').tail(10)
```
