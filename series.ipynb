```python
#integer
marks = [88 , 32 , 89 , 99 , 64 , 55 , 73]

pd.Series(marks)
```

```python
names = ["joy" , "roy" , "moi" , "loi" , "yoi" , "koi" , "toi"]
pd.Series(marks , index= names)
```
